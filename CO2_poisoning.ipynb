```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(open("CO2data.csv", 'r'))
df.head()
```

### Scores over time

```python
plt.scatter(df.DATE, df["RELATIVE SCORE"], s=4)
plt.xticks([df.DATE.iloc[0], df.DATE.iloc[-1]])
plt.xlabel("Date")
plt.ylabel("Score (% of average)")
plt.title("Score over time")
plt.show()
```

### Score vs. CO2

```python
plt.scatter(df.CO2, df["RELATIVE SCORE"], s=4)
plt.xlabel("CO2 (ppm)")
plt.ylabel("Score (% of average)")
plt.title("Score over time")
plt.show()
```

### Score vs. CO2, with time

```python
dates = pd.to_datetime(df.DATE)
days_since_start = [i.days for i in (dates-dates.iloc[0])]
plt.scatter(df.CO2, df["RELATIVE SCORE"], s=4, c = days_since_start)
plt.xlabel("CO2 (ppm)")
plt.ylabel("Score (% of average)")
plt.title("Score over time")
cbar = plt.colorbar()
cbar.ax.set_title("Day #")
plt.show()
```

### Scores over time, with CO2

```python
no_co2_df = df[pd.isnull(df.CO2)]
plt.scatter(no_co2_df.DATE, no_co2_df["RELATIVE SCORE"], s=2, color = 'red')
plt.scatter(df.DATE, df["RELATIVE SCORE"], s=4, c = df.CO2)
plt.xticks([df.DATE.iloc[0], df.DATE.iloc[-1]])
plt.xlabel("Date")
plt.ylabel("Score (% of average)")
cbar = plt.colorbar()
cbar.ax.set_title("CO2 (ppm)")
plt.title("Score over time")
plt.show()
```
